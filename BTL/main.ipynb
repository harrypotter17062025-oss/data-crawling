```python
# ==============================
# 1. IMPORT THƯ VIỆN
# ==============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# 2. ĐỌC FILE EXCEL
# ==============================
file = "MAT16A_Du_lieu_Tai_chinh_cong_Viet_Nam_2005_2020.xlsx"

# Xem tên các sheet
xls = pd.ExcelFile(file)
print("Danh sách sheet:", xls.sheet_names)

# ===============================
# 3. Xay dung dashboard
# ===============================

# Đọc dữ liệu từ sheet "Data_2005_2020"
df_master_data = pd.read_excel(file, sheet_name="Data_2005_2020", header=7)
#df_master_data = df_master_data.iloc[8:].reset_index(drop=True)
print(df_master_data.head())
```

```python

import pandas as pd

# ===== Lấy dữ liệu =====
df_year = df_master_data.iloc[:, 0]
df_gdp = pd.to_numeric(df_master_data.iloc[:, 1], errors='coerce')

# ===== Thống kê =====
gdp_mean = round(df_gdp.mean(), 2)
print("Trung bình tăng trưởng GDP:", gdp_mean)

max_gdp = round(df_gdp.max(), 2)
year_max = df_year.iloc[df_gdp.idxmax()]
print("Năm GDP cao nhất:", year_max, "với giá trị:", max_gdp)

min_gdp = round(df_gdp.min(), 2)
year_min = df_year.iloc[df_gdp.idxmin()]
print("Năm GDP thấp nhất:", year_min, "với giá trị:", min_gdp)

# ===== Vẽ biểu đồ đẹp hơn =====
plt.figure()

# Đường chính
plt.plot(df_year, df_gdp, marker='o')

# Highlight điểm max
plt.scatter(year_max, max_gdp)
plt.text(year_max, max_gdp, f" Max: {max_gdp} ({year_max})")

# Highlight điểm min
plt.scatter(year_min, min_gdp)
plt.text(year_min, min_gdp, f" Min: {min_gdp} ({year_min})")

# Đường trung bình
plt.axhline(gdp_mean, linestyle='--')
plt.text(df_year.iloc[0], gdp_mean, f" Mean: {gdp_mean}")

# Tiêu đề & label
plt.title("Tăng trưởng GDP theo năm (2005–2020)")
plt.xlabel("Năm")
plt.ylabel("GDP Growth (%)")

plt.grid()
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# ===== Lấy dữ liệu =====
df_year = df_master_data.iloc[:, 0]
df_balance = pd.to_numeric(df_master_data.iloc[:, 5], errors='coerce')

# ===== Tô màu =====
colors = ['green' if v >= 0 else 'red' for v in df_balance]

# ===== Vẽ biểu đồ =====
plt.figure()

bars = plt.bar(df_year, df_balance, color=colors)

# ===== Gắn % lên từng cột =====
for bar, value in zip(bars, df_balance):
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # vị trí ngang giữa cột
        value,                              # vị trí theo chiều cao
        f"{value:.2f}%",                    # format %
        ha='center',
        va='bottom' if value >= 0 else 'top'  # tránh bị đè vào cột
    )

# ===== Đường mốc 0 =====
plt.axhline(0)

# ===== Tiêu đề =====
plt.title("Cân đối ngân sách so với GDP (2005–2020)")
plt.xlabel("Năm")
plt.ylabel("Tỷ lệ (% GDP)")

plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# ===== Lấy dữ liệu =====
df_year = df_master_data.iloc[:, 0]
df_gdp_current_usd = pd.to_numeric(df_master_data.iloc[:, 6], errors='coerce')

# ===== Khởi tạo khung hình =====
plt.figure(figsize=(20, 8)) # Tăng kích thước để dễ nhìn hơn

# ===== Vẽ biểu đồ cột (Bar) =====
colors = ['green' if v >= 0 else 'red' for v in df_gdp_current_usd]
bars = plt.bar(df_year, df_gdp_current_usd, color=colors, alpha=0.6, label='GDP per Year')

# ===== Vẽ biểu đồ đường (Line) =====
# marker='o' tạo các chấm tại điểm dữ liệu, color='blue' để làm nổi bật đường kẻ
plt.plot(df_year, df_gdp_current_usd, color='darkblue', marker='o', linewidth=2, label='Trendline')

# ===== Gắn nhãn giá trị lên từng cột =====
for bar, value in zip(bars, df_gdp_current_usd):
    plt.text(
        bar.get_x() + bar.get_width() / 2, 
        value,                              
        f"{value:.2f}",                    
        ha='center',
        va='bottom' if value >= 0 else 'top',
        fontsize=10,
        fontweight='bold'
    )

# ===== Định dạng biểu đồ =====
plt.axhline(0, color='black', linewidth=1) # Đường mốc 0 đậm hơn tí
plt.title("GDP Hàng Năm Theo USD (2005–2020)", fontsize=16)
plt.xlabel("Năm", fontsize=12)
plt.ylabel("Giá trị (Tỷ USD)", fontsize=12)

plt.xticks(df_year, rotation=45) # Đảm bảo hiển thị đủ các năm trên trục X
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend() # Hiển thị chú thích cho cột và đường

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ===== 1. Lấy dữ liệu =====
df_year = df_master_data.iloc[:, 0]
df_gdp_revenue_usd = pd.to_numeric(df_master_data.iloc[:, 7], errors='coerce')
df_gdp_expenditure_usd = pd.to_numeric(df_master_data.iloc[:, 8], errors='coerce')

# ===== 2. Thiết lập thông số hiển thị =====
x = np.arange(len(df_year))  
width = 0.4  # Tăng độ rộng cột một chút để nhìn cho đầy đặn

# ===== 3. Khởi tạo khung hình SIÊU LỚN =====
# figsize=(25, 10) giúp biểu đồ giãn ra rất rộng theo chiều ngang
plt.figure(figsize=(25, 10), dpi=100) 

# ===== 4. Vẽ biểu đồ cột đôi =====
bars1 = plt.bar(x - width/2, df_gdp_revenue_usd, width, label='Thu ngân sách', color='#1f77b4', edgecolor='white')
bars2 = plt.bar(x + width/2, df_gdp_expenditure_usd, width, label='Chi ngân sách', color='#ff7f0e', edgecolor='white')

# ===== 5. Hàm gắn chữ lên đầu cột (Cỡ chữ to hơn) =====
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        # va='bottom' đẩy chữ lên cao hơn đỉnh cột một chút để tránh dính
        plt.text(
            bar.get_x() + bar.get_width() / 2, 
            height + (max(df_gdp_revenue_usd)*0.01), # Tạo khoảng cách nhỏ với đỉnh cột
            f"{height:.1f}",
            ha='center', 
            va='bottom',
            fontsize=12,      # Cỡ chữ số trên cột
            fontweight='bold',
            rotation=0        # Để chữ nằm ngang cho dễ đọc
        )

autolabel(bars1)
autolabel(bars2)

# ===== 6. Định dạng chữ và khoảng cách =====
plt.axhline(0, color='black', linewidth=1.5)

# Tiêu đề cực to
plt.title("SO SÁNH THU - CHI NGÂN SÁCH (2005–2020)", fontsize=24, fontweight='bold', pad=30)

# Nhãn trục X, Y to rõ
plt.xlabel("Năm", fontsize=16, labelpad=15)
plt.ylabel("Giá trị (Tỷ USD)", fontsize=16, labelpad=15)

# Tọa độ trục X giãn cách đều
plt.xticks(x, df_year, rotation=0, fontsize=14) # Để rotation=0 nếu muốn chữ nằm ngang
plt.yticks(fontsize=14)

# Chú thích (Legend)
plt.legend(fontsize=15, loc='upper left', frameon=True)

# Lưới mờ giúp dễ đối chiếu
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Tự động căn chỉnh để không mất chữ ở rìa
plt.tight_layout()

plt.show()
```
